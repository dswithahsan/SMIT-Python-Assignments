```python
data["Death_Rate"] = (data["Deaths"] / data["Total_cases"]) * 100
```

```python
low_death_rate_countries = data[(data["Total_cases"] > data["Confirmed last week"]) & (data["Death_Rate"] < 2)]
print(low_death_rate_countries[["Country/Region", "Total_cases", "Death_Rate"]])

```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(10, 6))
plt.scatter(data["Total_cases"], data["Deaths"], alpha=0.5)
plt.xlabel("Total Cases")
plt.ylabel("Total Deaths")
plt.title("Scatter Plot: Total Cases vs. Total Deaths")
plt.xscale("log")  # Log scale for better visualization if numbers are large
plt.yscale("log")
plt.grid(True)
plt.show()
```
